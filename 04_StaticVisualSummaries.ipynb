# Visual summaries
As well as being able to analyse and summarise our data, we often need to visualise it. This section looks at the key data visualisation package `ggplot2` and looks at themes made available from different sources like the NYT in the `ggthemes` package.

```r
install.packages(c("tidyverse","ggthemes"))
library(tidyverse)
heroes = read_csv("clean/heroes.csv")
powers = read_csv("clean/powers_long.csv")
```

We can make plots using our data but we need a few terms to translate how we think of Excel charts to code first.

- `ggplot()`: A chart in Excel is a plot in R
- `aes()`: A series or something else being represented is an aesthetic
- `geom_*()`: A chart type is a geometry
- `theme()`: Formatting is theming

We get a series of R functions that we can add together to make a whole chart. The really nifty thing is that because these are are code, we can make them into reusable components so we can build on them, reuse them, and even use the same chart with different data. It's incredibly handy when you need multiple charts.

```r
ggplot(heroes, aes(x=Alignment)) +
  geom_bar()
```

We can do substantial theming and customisation so it isn't just a basic chart, including using themes from the `ggthemes` package.

```r
library(ggthemes)
ggplot(heroes, aes(x=Alignment)) +
  geom_bar() +
  theme_economist() 
```

```r
ggplot(heroes, aes( x = Height, y = Weight, colour = Alignment )) +
  geom_point(alpha=.5) +
  theme_wsj() + 
  scale_colour_wsj()
```
